```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```
df = pd.read_csv("SampleSuperstore.csv")
df.head()
```

```

from sklearn.preprocessing import LabelEncoder
```

```
le = LabelEncoder()
for col in df.columns:
    df[col] = le.fit_transform(df[col])
```

```
sns.heatmap(df.corr(),robust=True)
```

```

sns.barplot(df['Ship Mode'],df['Segment'])
```

```
sns.barplot(df['City'],df['State'])
```

```
sns.barplot(df['Postal Code'],df['Region'])
```

```
sns.countplot(df['Sales'])
```

```
sns.countplot(df['Quantity'])
```

```

sns.boxplot(df)
```

```
sns.clustermap(df)
```

```

sns.lvplot(df)
```

```

data = df.iloc[:40,:]
sns.pairplot(data)
```

```
sns.violinplot(df['Discount'],df['Profit'])
```

```
sns.stripplot(df['Discount'],df['Profit'])
```
